```python
import random

# Label genre seni
genre_seni_label = {'Abstract': 1,
                    'Cubism': 2,
                    'Dadaism': 3,
                    'Fauvism': 4,
                    'Impressionism': 5,
                    'Nouveau': 6,
                    'Pop': 7,
                    'Realism': 8,
                    'Renaissance': 9,
                    'Surrealism': 10,
                    }

genres = list(genre_seni_label.keys())

# Generate dataset secara acak
dataset = []
for i in range(100):
    user_id = i + 1
    nama = f"User_{user_id}"

    # Generate genre seni secara acak untuk user
    num_genres = random.randint(1, len(genres))
    user_genres = random.sample(genres, num_genres)

    # Generate rating secara acak untuk setiap genre, dengan minimal rating 1
    ratings = [random.randint(1, 10) for _ in range(num_genres)]

    # Menambahkan setiap data ke dataset
    for genre, rating in zip(user_genres, ratings):
        genre_id = genre_seni_label[genre]  # Ambil ID dari label genre
        # Tambahkan ID genre ke dalam dataset
        dataset.append([user_id, nama, genre, genre_id, rating])

# Print dataset
print("user_id, nama, genre_seni, genre_seni_id, rating")
for data in dataset:
    print(", ".join(map(str, data)))
```

```python
import csv
import os

# Direktori untuk menyimpan file CSV
directory = "..\Dataset"
if not os.path.exists(directory):
    os.makedirs(directory)

# Lokasi file CSV
csv_file = os.path.join(directory, "dataset_dummy.csv")

# Tulis dataset ke dalam file CSV
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    # Menulis header
    writer.writerow(["user_id", "nama", "genre_seni",
                    "genre_seni_id", "rating"])
    # Menulis setiap baris dataset
    writer.writerows(dataset)

print("Dataset telah disimpan dalam file CSV:", csv_file)
```
